```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) |
          (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
skewed_cols = df[num_cols].skew()[df[num_cols].skew() > 1].index

for col in skewed_cols:
    df[col] = np.log1p(df[col])
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
skewness_values = df[num_cols].skew()
for col, value in skewness_values.items():
    if value > 0:
        print(f"{col}: Positively skewed (Right-skewed)")
    elif value < 0:
        print(f"{col}: Negatively skewed (Left-skewed)")
    else:
        print(f"{col}: Symmetrical distribution")
```


## **EDA and Visualizations**

**Univariate Analysis** → distribution of single variables (countplot, histogram, boxplot)


```python
#1. Categorical Variable – Count Plot
#Gender Distribution

plt.figure()
sns.countplot(x='Gender', data=df)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### **Insights**

* Shows the number of patients by gender

* Helps identify class imbalance in categorical variables

### **Interpretation**

* This chart shows the gender composition of patients visiting the hospital.

* A dominant gender group indicates higher healthcare utilization by that segment.

* This insight helps hospital management plan gender-specific facilities, staffing, and awareness programs.

```python
#2.Numerical Variable – Histogram
# Age Distribution

plt.figure()
plt.hist(df['Age'], bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### **Insights**

* Reveals age concentration and spread

* Identifies skewness in age distribution

### **Interpretation**

* The age distribution highlights the most frequent patient age groups.

* A higher concentration in middle or senior age groups indicates increased demand for chronic and long-term care services.

* Hospitals can use this insight to improve specialist availability and preventive care programs.

```python
#3. Numerical Variable – Box Plot
#Billing Amount Distribution

plt.figure()
plt.boxplot(df['Billing Amount'])
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

### **Insights**

* Identifies median, quartiles, and outliers

* Helps detect extreme hospital stay durations

### **Interpretation**

* The median hospital stay reflects the typical resource usage per patient.

* Outliers indicate patients with unusually long stays, which increase operational costs.

* Identifying such cases helps hospitals improve bed management and discharge planning.

## **EDA and Visualizations**


**Bivariate Analysis** → relation between two variables (scatterplot, barplot, correlation heatmap)

```python
#1. Scatter Plot (Numerical vs Numerical)
#Age vs Billing Amount

plt.figure()
plt.scatter(df['Age'], df['Billing Amount'])
plt.title('Age vs Billing Amount')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.show()
```

### **Insights**

* Helps understand how billing amount varies with patient age

* Identifies trends, clusters, or anomalies

### **Interpretation**

* This plot shows how medical expenses vary across age groups.

* Higher billing amounts among older patients suggest increased treatment complexity.

* This insight supports insurance pricing, senior care packages, and cost forecasting.

```python
#2. Bar Plot (Categorical vs Numerical)
#Average Billing Amount by Gender

plt.figure()
sns.barplot(x='Gender', y='Billing Amount', data=df)
plt.title('Average Billing Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Billing Amount')
plt.show()
```

### **Insights**

* Compares average billing across categories

* Highlights cost differences among patient groups

### **Interpretation**

* The chart compares healthcare spending across genders.

* Noticeable differences suggest variation in treatment needs or duration.

* This helps management design gender-aware budgeting and healthcare plans.

```python
#Correlation Heatmap (Numerical vs Numerical)

plt.figure()
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

### **Insights**

* Shows strength and direction of relationships

* Helps identify highly correlated variables

* Supports feature selection and modeling

### **Interpretation**

* Strong correlation between hospital stay days and billing amount indicates cost increases with longer stays.

* Weak correlation with age suggests age alone does not drive costs.

* This insight helps focus cost-reduction strategies on operational efficiency rather than demographics.

## **EDA and Visualizations**

**Multivariate Analysis** → relation among 3+ variables (pairplot, grouped analysis, pivot tables, advanced plots)

```python
#1. Pair Plot (Multiple Numerical Variables)

sns.pairplot(df[['Age', 'Date of Admission', 'Billing Amount']], diag_kind='kde')
plt.show()
```

### **Insights**

* Shows pairwise relationships between numerical variables

* Helps detect correlations, clusters, and outliers

* Diagonal plots display individual variable distributions

### **Interpretation**

* The pair plot confirms that billing amount rises with hospital stay duration.

* Wide dispersion indicates that treatment type and severity also affect cost.

* This insight supports predictive cost modeling and patient risk assessment.

```python
#2. Grouped Analysis (Categorical + Multiple Numerical Variables)
#Average Metrics by Gender

grouped_gender = df.groupby('Gender')[['Billing Amount', 'Age']].mean()
grouped_gender
```

### **Insights**

* Compares multiple healthcare outcomes across gender

* Helps identify demographic-based patterns

```python
#3. Pivot Table Analysis
#Average Billing by Treatment Type and Gender

pivot_table = pd.pivot_table(
    df,
    values='Billing Amount',
    index='Medication',
    columns='Gender',
    aggfunc='mean'
)
pivot_table
```

### **Insights**

* Reveals combined impact of treatment type and gender on billing

* Useful for operational and cost analysis

### **Interpretation**

* This table shows which treatment types are most expensive across genders.

* High-cost treatments highlight areas for cost optimization and supplier negotiation.

* Useful for strategic pricing and service planning.

```python
#4. Advanced Visualization (Grouped Bar Plot)
#Billing Amount by Treatment Type and Gender

plt.figure()
sns.barplot(x='Medication', y='Billing Amount', hue='Gender', data=df)
plt.title('Billing Amount by Treatment Type and Gender')
plt.xlabel('Treatment Type')
plt.ylabel('Billing Amount')
plt.xticks(rotation=45)
plt.show()
```

### **Interpretation**

* The chart visually compares treatment costs across genders.

* Some treatments consistently show high costs regardless of gender.

* Helps management prioritize high-cost departments for efficiency improvement.
